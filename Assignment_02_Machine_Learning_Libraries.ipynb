#1. Create a NumPy array of shape (5, 5) with random integers between 1 and 50.
#o Find the row with the maximum sum.
#o Replace all elements greater than the mean with 0.

```python
import numpy as np

np.random.seed(42)  # for reproducibility
arr = np.random.randint(1, 51, size=(5, 5))
arr
```

### Find the row with the maximum sum

```python
row_sums = arr.sum(axis=1)
print(row_sums)
max_sum_row_index = np.argmax(row_sums)
max_sum_row = arr[max_sum_row_index]
max_sum_row
```

### Replace all elements greater than the mean with 0.

```python
mean_value = arr.mean()
arr_modified = np.where(arr > mean_value, 0, arr)

mean_value, arr_modified
```

# 3. Load the dataset Iris data
# o Display the first 10 rows and basic statistics.
# o Filter all rows where sepal_length > median and species is "versicolor".
# o Group the data by species and compute the mean petal_length.

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

df.head()
```

### Display first 10 rows and basic statistics

```python

df.head(10)
```

```python

df.describe()
```

### Filter all rows where sepal_length > median and species is "versicolor".


```python
median_sepal_length = df['sepal length (cm)'].median()

filtered_df = df[
    (df['sepal length (cm)'] > median_sepal_length) &
    (df['species'] == 'versicolor')
]
filtered_df
```

### Group the data by species and compute the mean petal_length.

```python

grouped_mean = df.groupby('species')['petal length (cm)'].mean()
grouped_mean
```

# 4. Add a new column named petal_ratio = petal_length / petal_width.
# o Find the top 5 rows with the highest petal_ratio.

```python
df['petal_ratio'] = df['petal length (cm)'] / df['petal width (cm)']
df.head()
```

### Find the top 5 rows with the highest petal_ratio.

```python
df.sort_values(by='petal_ratio', ascending=False).head(5)
```

# 5. Using Matplotlib, create:
# o A histogram of sepal_length with 15 bins.
# o A scatter plot of sepal_length vs petal_length, color-coded by species.  

```python
import matplotlib.pyplot as plt

#A histogram of sepal_length with 15 bins.
plt.figure()
plt.hist(df['sepal length (cm)'], bins=15)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

### A scatter plot of sepal_length vs petal_length, color-coded by species.

```python

plt.figure()

for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(
        subset['sepal length (cm)'],
        subset['petal length (cm)'],
        label=species
    )

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Sepal Length vs Petal Length')
plt.legend()
plt.show()
```
