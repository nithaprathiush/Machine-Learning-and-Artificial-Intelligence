#1. Simple ATM Calculator

```python

balance = int(input("Enter account balance: "))
withdrawal = int(input("Enter withdrawal amount: "))

if withdrawal <= balance:
    balance -= withdrawal
    print(f"Transaction successful. Remaining balance: ₹{balance}")
else:
    print("Insufficient balance. Transaction failed.")
```

#2. Electricity Bill Generator

```python

units = int(input("Enter units consumed: "))

bill = 0

if units <= 100:
    bill = units * 5
elif units <= 200:
    bill = (100 * 5) + (units - 100) * 7
else:
    bill = (100 * 5) + (100 * 7) + (units - 200) * 10

print(f"Bill amount = ₹{bill}")
```

#3. Temperature Converter

```python

temp = float(input("Enter temperature value: "))
unit = input("Enter unit (C/F): ").upper()

if unit == "C":
    fahrenheit = (temp * 9/5) + 32
    print(f"Temperature in Fahrenheit: {fahrenheit}°F")
elif unit == "F":
    celsius = (temp - 32) * 5/9
    print(f"Temperature in Celsius: {celsius}°C")
else:
    print("Invalid unit entered.")
```

#4. Mobile Recharge Validity

```python
print("=== Mobile Recharge Validity ===")

from datetime import datetime

recharge_date = input("Enter recharge date (YYYY-MM-DD): ")
today_date = input("Enter today's date (YYYY-MM-DD): ")

recharge = datetime.strptime(recharge_date, "%Y-%m-%d")
today = datetime.strptime(today_date, "%Y-%m-%d")

days_used = (today - recharge).days

if days_used <= 28:
    print("Plan is still valid.")
else:
    print("Plan expired. Recharge again.")
```

#6. Sum of Even Numbers

```python

num = int(input("Enter a number: "))
even_sum = 0

for i in range(2, num + 1, 2):
    even_sum += i

print(f"Sum of even numbers: {even_sum}")
```

#7. Dictionary – Student Marks Summary

```python

students = {}
total_marks = 0

for i in range(3):
    name = input("Enter name: ")
    marks = int(input("Enter marks: "))
    students[name] = marks
    total_marks += marks

average = total_marks / 3
print(f"Average Marks: {average}")
```

#8. Dictionary – Word Frequency

```python

sentence = input("Enter a sentence: ")
words = sentence.split()

frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

for word, count in frequency.items():
    print(f"{word}: {count}")
```

#8. A. Functions – Check Prime

```python

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))

if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

#9. Vowel Counter

```python

def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

string = input("Enter a string: ")
print(f"Number of vowels: {count_vowels(string)}")
```

#10. Count Divisible Numbers

```python

N = int(input("Enter the range limit (N): "))
D = int(input("Enter the divisor (D): "))

count = 0

for i in range(1, N + 1):
    if i % D == 0:
        count += 1

print(f"Count of numbers divisible by {D}: {count}")
```
